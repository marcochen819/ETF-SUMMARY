```python
import requests
import pandas as pd
import time
import datetime
import random

def get_twse_etf_data(date_str):
    url = f"https://www.twse.com.tw/rwd/zh/fund/T86?response=json&date={date_str}&selectType=0099P"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }

    try:
        resp = requests.get(url, headers=headers, timeout=10)
        data = resp.json()

        if data.get("stat") == "OK":
            df = pd.DataFrame(data["data"], columns=data["fields"])
            selected_cols = ["證券代號", "證券名稱", "外陸資買賣超股數(不含外資自營商)",
                             "投信買賣超股數", "自營商買賣超股數", "三大法人買賣超股數"]
            df = df[selected_cols]
            df.insert(0, '日期', date_str)
            return df
    except Exception as e:
        print(f"日期 {date_str} 抓取失敗: {e}")
    return None

# --- 自行設定爬蟲區間 ---
start_date = "2026-01-01"
end_date = "2026-01-31"

# 產生日期序列 (自動過濾掉週六、週日)
# 確保日期格式為純數字 20260105，不要有橫線
date_list = pd.date_range(start=start_date, end=end_date, freq='B').strftime("%Y%m%d")

all_dfs = []

for d in date_list:
    print(f"正在抓取 {d}...")
    res = get_twse_etf_data(d)

    if res is not None:
        all_dfs.append(res)
        print(f"✅ {d} 成功")
    else:
        print(f"❌ {d} 無資料 (可能是休市)")

    # 在 3 到 6 秒之間隨機取一個數字
sleep_time = random.uniform(3, 6)
time.sleep(sleep_time) # 務必保留，避免被封鎖 IP

# 合併輸出
if all_dfs:
    final_result = pd.concat(all_dfs, ignore_index=True)
    final_result.to_csv("ETF_Summary.csv", encoding="utf-8", index=False)
    print("\n任務完成！檔案已儲存。")
else:
    print("\n沒有抓到任何資料。")
```
